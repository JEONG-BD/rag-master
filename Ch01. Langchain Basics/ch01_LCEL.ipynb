```python
prompt_template = '주제 {topic}에 대해서 짧은 설명을 해주세요' 
```

```python
def call_chat_model(message: List[dict]):
    response = client.chat.completions.create(messages=message, model='google/gemma-4-e4b') 
    print(response.choices[0].message.content)
    print(dir(response)) 

def invoke_chain(topic: str):
    prompt_value = prompt_template.format(topic=topic)
    messages = [{'role' : 'user' , 'content' : prompt_value}]
    return call_chat_model(messages)

invoke_chain('더블딥')
```

```python
from langchain_openai import ChatOpenAI 
from langchain_core.prompts import ChatPromptTemplate 
from langchain_core.output_parsers import StrOutputParser 
from langchain_core.runnables import RunnablePassthrough 
```

```python
prompt = ChatPromptTemplate.from_template(
    '주제 {topic}에 대해 짧은 설명을 해주세요'
)
```

```python

llm = ChatOpenAI(
    model="gemma-3-4b-it",
    base_url="http://localhost:1234/v1",
    api_key="lm-studio"
)
```

```python
output_parser =StrOutputParser()
```

```python
chain =(
    {'topic' : RunnablePassthrough()} 
    | prompt 
    | llm 
    | output_parser
)
```

```python
chain.invoke('더블딥')
```

#### 배치 출력 

```python
chain.batch([{'topic':'더블딥'}, {'topic':'인플레이션'}])
```

#### 스트림 출력

```python
for token in chain.stream({'topic': '더블딥'}):
    print(token, end='', flush=True)
```

#### 러너블을 체인으로 연결

```python
from langchain_core.prompts import ChatPromptTemplate 
from langchain_core.output_parsers import StrOutputParser 

analysis_pormpt = ChatPromptTemplate.from_template('이 대답을 영어로 번역해주세요 {answer}')
```

```python
composed_chain = {'answer' : chain} | analysis_pormpt | llm | StrOutputParser()
```

```python
composed_chain.invoke({'topic': '더블딥'})
```

#### 람다 함수를 사용한 체인을 구성하기

```python
prompt = ChatPromptTemplate.from_template("{topic}에 대해 짧게 설명을 해주세요.")
chain = prompt | llm | StrOutputParser()
analysis_prompt = ChatPromptTemplate.from_template("이 대답을 영어로 번역해 주세요: {answer}")

composed_chain_with_lambda = (
    chain
    | (lambda input: {"answer": input})
    | analysis_prompt
    | llm
    | StrOutputParser()
)
composed_chain_with_lambda.invoke({"topic": "더블딥"})
```

#### 파이프 메서드

```python
composed_chain_with_pipe = (
    chain.pipe(lambda input : {'answer' : input})
    .pipe(analysis_pormpt)
    .pipe(llm)
    .pipe(StrOutputParser())
)

composed_chain_with_pipe.invoke({'topic': '더블딥'})
```

```python
composed_chain_with_pipe = chain.pipe(lambda input : {'answer' : input}, analysis_pormpt, 
                                      llm, StrOutputParser())

composed_chain_with_pipe.invoke({'topic':'더블딥'})
```

#### RunnableParallel 체인 구성


```python
from langchain_core.runnables import RunnableParallel 

kor_chain = (
    ChatPromptTemplate.from_template('{topic}에 대해 짧은 설명을 해주세요')
    | llm 
    | StrOutputParser()
)

eng_chain = (
    ChatPromptTemplate.from_template('{topic}에 대해 짧게 영어로 설명을 해주세요')
    | llm 
    | StrOutputParser()
)

parallel_chain = RunnableParallel(kor=kor_chain, eng=eng_chain)
result = parallel_chain.invoke({'topic': '더블딥'})

print('한글 설명', result['kor'])
print('영어 설명', result['eng'])
```
